# **Data Preprocessing**

## 1. Handling categorical values
  #### Encoding string values to integer constants using LabelEncoder 

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
dataset.iloc[:,2]=lab.fit_transform(dataset.iloc[:,2])
```

#### Dropping unwanted columns

```python
dataset.drop(["Unnamed: 0", "id", "date"],axis = 1, inplace = True)
```

# **Slicing the dataset to separate features from prediction**

```python
X = (dataset.drop('price', axis=1)).to_numpy()
y = dataset.iloc[:, 0].values
```

# **Checking the dataset for null values**

```python
dataset.isnull().sum()
```

## 2. Handling missing values
  #### Filling missing values using MEAN with the help of SimpleImputer class of scikit-learn  

```python
from sklearn.impute import SimpleImputer
sim1 = SimpleImputer(missing_values = np.nan, strategy = 'mean')
sim1 = sim1.fit(X[:, [1]])
X[:, [0]] = sim1.transform(X[:, [0]])

sim2 = SimpleImputer(missing_values = np.nan, strategy = 'mean')
sim2 = sim1.fit(X[:, [2]])
X[:, [1]] = sim1.transform(X[:, [1]])
```

## 3. Performing feature scaling
#### Standardization using StandardScaler class to bring observations on same scale, i.e., between a range

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X = pd.DataFrame(X)
X.head()
```

```python
X.isnull().sum()
```

## 4. Outlier detection and correction

#### Detecting outliers in the dataset using IQR Score

```python
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
IQR_1 = ((dataset < (Q1 - 1.5 * IQR))  | (dataset > (Q3 + 1.5 * IQR)))
IQR_1.head()
```

#### Removing outliers from the dataset using IQR Score

```python
dataset = dataset[~((dataset < (Q1-1.5*IQR)) | (dataset > (Q3+1.5*IQR))).any(axis=1)]
dataset.shape
```

# **Importing the necessary module and splitting the dataset into training and testing set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# **Model development**

## 1. Implementing Simple Linear Regression as this is a regression problem because here the output variable is of continuous nature

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

## Fitting the lin_reg object on the training set using fit() and calculating the value of coeffifcient and intercept 

```python
lin_reg.fit(X_train, y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

## Calculating the accuracy of model using score() function

```python
lin_reg.score(X_train, y_train) 
```

```python
lin_reg.score(X_test, y_test) 
```

#### As the accuracy of training set is 70% and testing set is 68%, i.e., there is not much difference. Thus, model is not overfitting

## Making the prediction on training data using predict()

```python
y_train_pred = lin_reg.predict(X_train)
y_train_pred
```

## Using performance metrices to evaluate model's performance on training set

#### **1. RMSE(Root Mean Square Error):**It is a standard way to measure error of a model. Its value increases exponentially if data has outliers.

```python
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))
print(RMSE)
```

#### **2. MAE(Mean Absolute Error):**If data contains a lot of outliers then MAE is preferred over RMSE.

```python
MAE = mean_squared_error(y_train, y_train_pred)
print(MAE)
```

#### **3.Coefficient of determination(R^2):**score() function calculates value of R^2. It always outputs a value between limits of [0 to 1]. Closer to 1 value specifies good behaviour of model and closer to 0 specifies poor model.

```python
from sklearn.metrics import r2_score
r = r2_score(y_train, y_train_pred)
print(r)
```

#### **4. Adjusted R^2:**This metric has been smartly derived to punish model for adding irrelevant features by lowering its value. Whereas it also rewards model for adding relevant features or strong predictors.

```python
adj_r = 1 - (1-r)*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1)
print(adj_r)
```

## Making the prediction on test data using predict()

```python
y_test_pred = lin_reg.predict(X_test)
y_test_pred
```

## Similarly using performance metrices to evaluate model's performance on test set

```python
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
print(RMSE)
```

```python
MAE = mean_squared_error(y_test, y_test_pred)
print(MAE)
```

```python
from sklearn.metrics import r2_score
r = r2_score(y_test, y_test_pred)
print(r)
```

```python
adj_r = 1 - (1-r)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print(adj_r)
```
